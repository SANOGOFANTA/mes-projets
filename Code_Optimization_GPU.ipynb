```
import cudf
import pandas as pd
import cupy as cp
from cuml.neighbors import KNeighborsClassifier
from cuml.model_selection import train_test_split
from cuml.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from cuml.preprocessing import LabelEncoder
import time
start_time = time.time()

# Charger les données avec cuDF
df = cudf.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/loan_train.csv')

# Convertir temporairement en pandas pour nettoyer les dates
df_pandas = df.to_pandas()
df_pandas['due_date'] = pd.to_datetime(df_pandas['due_date'], errors='coerce')  # Normaliser les dates
df_pandas['effective_date'] = pd.to_datetime(df_pandas['effective_date'], errors='coerce')

# Revenir à cuDF après nettoyage
df = cudf.DataFrame(df_pandas)

# Vérification des premières lignes
print(df.head())

# Histogrammes avec seaborn (conversion cuDF -> pandas pour compatibilité)
df_pandas = df.to_pandas()

# Visualisation des données
bins = cp.linspace(df['Principal'].min(), df['Principal'].max(), 10).get()
g = sns.FacetGrid(df_pandas, col="Gender", hue="loan_status", palette="Set1", col_wrap=2)
g.map(plt.hist, 'Principal', bins=bins, ec="k")
g.axes[-1].legend()
plt.show()

bins = cp.linspace(df['age'].min(), df['age'].max(), 10).get()
g = sns.FacetGrid(df_pandas, col="Gender", hue="loan_status", palette="Set1", col_wrap=2)
g.map(plt.hist, 'age', bins=bins, ec="k")

g.axes[-1].legend()
plt.show()

# Ajouter les jours de la semaine
df['dayofweek'] = df['effective_date'].dt.weekday

# **Update df_pandas after adding the 'dayofweek' column**
df_pandas = df.to_pandas() # This line has been added

bins = cp.linspace(df['dayofweek'].min(), df['dayofweek'].max(), 10).get()
g = sns.FacetGrid(df_pandas, col="Gender", hue="loan_status", palette="Set1", col_wrap=2)
g.map(plt.hist, 'dayofweek', bins=bins, ec="k")
g.axes[-1].legend()
plt.show()


# Utiliser map() pour effectuer le remplacement des valeurs
df['Gender'] = df['Gender'].map({'male': 0, 'female': 1})

# Vérification des résultats
print(df['Gender'].head())

df['weekend'] = df['dayofweek'].apply(lambda x: 1 if (x>3)  else 0)
print(df.head())
# Sélectionner les features de base
Feature = df[['Principal','terms','age','Gender','weekend']]

# Ajouter les variables catégorielles via one-hot encoding (using cuDF)
Feature = cudf.concat([Feature, cudf.get_dummies(df['education'])], axis=1)

# Supprimer la colonne 'Master or Above'
Feature.drop(['Master or Above'], axis=1, inplace=True)

# Préparation de la cible
y = df['loan_status']
X=Feature

# Convertir les colonnes booléennes en int (0 ou 1)
bool_cols = X.select_dtypes(include=['bool']).columns
X[bool_cols] = X[bool_cols].astype(int)

# Vérification
print(X.head())
print("norm")
# Define numeric_cols before using it
numeric_cols = X.select_dtypes(include=['number']).columns

# Appliquer la normalisation
X[numeric_cols] = (X[numeric_cols] - X[numeric_cols].mean()) / X[numeric_cols].std()

# Vérification
print(X.head())

y = df['loan_status']

# Convert string labels to numerical using LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

# Diviser le dataset en train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

print('Train set:', X_train.shape, y_train.shape)
print('Test set:', X_test.shape, y_test.shape)

# Déterminer la meilleure valeur de K
Ks = 10
mean_acc = cp.zeros(Ks - 1)
std_acc = cp.zeros(Ks - 1)

for n in range(1, Ks):
    # Entraînement du modèle et prédictions
    neigh = KNeighborsClassifier(n_neighbors=n)
    neigh.fit(X_train, y_train)
    yhat = neigh.predict(X_test)

    # Calcul de la précision
    mean_acc[n - 1] = accuracy_score(y_test, yhat)
    # Fix: Convert the cuDF Series to a NumPy array before calculating std
    std_acc[n - 1] = cp.std( (yhat == y_test).to_numpy() ) / cp.sqrt(yhat.shape[0])

# Graphique de précision
mean_acc = mean_acc.get()
std_acc = std_acc.get()
plt.plot(range(1, Ks), mean_acc, 'g')
plt.fill_between(range(1, Ks), mean_acc - 1 * std_acc, mean_acc + 1 * std_acc, alpha=0.10)
plt.fill_between(range(1, Ks), mean_acc - 3 * std_acc, mean_acc + 3 * std_acc, alpha=0.10, color="green")
plt.legend(('Accuracy', '+/- 1xstd', '+/- 3xstd'))
plt.ylabel('Accuracy')
plt.xlabel('Number of Neighbors (K)')
plt.tight_layout()
plt.show()
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1)
#Build the model this time using the k value that produced the highest accuracy
knn = KNeighborsClassifier(n_neighbors = mean_acc.argmax()+1)
#Fit the model with the training set
knn.fit(X_train, y_train)

#Make some predictions
knn_yhat = knn.predict(X_test)
#Make some predictions
knn_yhat = knn.predict(X_test)
print(knn_yhat[0:5])


end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")



```
